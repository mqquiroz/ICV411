## Ejercicios Clase 2

Calcule la pérdida por fricción en una tubería de hierro fundido de 500 m de largo $\epsilon = 0.26$ mm. El diámetro es de 30 cm, el caudal es de 0.02 m$^{3}$/s y se supone una temperatura de 20 °C ($\nu = 1 \times 10^{-6}$ m$^{2}$/s)

```python
#librerias
import numpy as np

#Datos
L  = 500  #Longitud (m)
e  = 0.26 #Rugosidad (mm)
D  = 30   #Diámetro (cm)
Q  = 0.02 #Caudal (m^3/2)
nu = 1e-6 #Viscosidad Cinemática (m^2/s)

#----------------
g = 9.81 #(m/s^2)
D = D/100 #(m)
e = e/1000 #(m)
```

```python
#Cáculos

#Área
A = np.pi*D**2/4 #(m^2)
#Velocidad
U = Q/A #(m/s)

Re = U*D/nu
print('Re =',np.round(Re,2))
e_D = e/D
print('e_D =',np.round(e_D,6))

#Swamee & Jain, 1976 aproximación
xx = (np.log10((e_D/3.7) + (5.74/(Re**0.9)) ))**2
f = 0.25/xx
print('f =',np.round(f,6))

dHf = f*(L/D)*(U**2/(2*g))
print('dHf =',np.round(dHf,6),'m')
```

### Ejercicio 3

Para la tubería de la figura, calcule la presión a mitad de camino entre A y B. ¿Cuál es la pérdida de energía en la válvula entre B y C? ¿Por qué la boquilla está al final de la tubería?

```python
#Datos

g     = 32.2 #(ft/s^2)
gamma = 62.4 #(lb/ft^3)

H    = 60 #(ft)
K_A  = 0.5
K_BC = 10
K_E  = 0.1
D    = 10 #(in)
D_E  = 3  #(in)
f    = 0.02
L_AB = 80 #(ft)
L_CD = 120 #(ft)

# ----------------
D = D/12 #(ft)
D_E = D_E/12 #(ft)
```

```python
Ltotal = L_AB + L_CD
print('Ltotal =',np.round(Ltotal,2),'ft')

xx = (f*(Ltotal/D)) + (K_A+K_BC) + (1.1*((D/D_E)**4))

U = np.sqrt(H*2*g/xx)
print('U =',np.round(U,2),'ft/s')
U_E = U*((D/D_E)**2)
print('U_E =',np.round(U_E,2),'ft/s')

U22g = ((U**2)/(2*g)) #(ft)

dH_BC = K_BC*U22g
print('dH_BC =',np.round(dH_BC,2),'ft')

dH_A = K_A*U22g
print('dH_A =',np.round(dH_A,2),'ft')

dHf_barAB = f*(0.5*L_AB/D)*U22g
print('dHf_barAB =',np.round(dHf_barAB,2),'ft')

dH_barAB = dHf_barAB + dH_A

P_barAB = gamma*(H - U22g - dH_barAB)
print('P_barAB =',np.round(P_barAB,2),'psf')
print('h_P_barAB =',np.round(P_barAB/gamma,2),'ft')
```
