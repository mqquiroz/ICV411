```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#declaramos un vector en el eje x

dx = 5
x = np.arange(0,100+dx,dx)

y = x

X,Y = np.meshgrid(x,y)

plt.plot(X,Y,'k*')
plt.plot(X.T,Y.T,'k*')
plt.show()
```

```python
Phi = X**2 + Y

plt.pcolormesh(X,Y,Phi)
plt.plot(50,50,'r*')
plt.colorbar()
plt.show()
```

```python
U = X + Y
V = Y**2 + X

plt.quiver(X,Y,U,V)
plt.show()
```

```python
plt.contour(X,Y,Phi,colors='k')
plt.show()
```
