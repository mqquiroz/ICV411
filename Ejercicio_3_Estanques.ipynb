El embalse $A$ abastece a dos turbinas del tipo de "pasada". Suponiendo que el embalse entrega un caudal constante de 3 m$^3$/s y que ambas turbinas funcionan con la misma altura neta, se pide determinar el caudal de operación de cada turbina. 


Considere:

Potencia total generada 5.880 kW

Turbinas 100% eficientes

Pérdida de energía de cada tubería dada por $\Delta H = R\cdot Q^{2}$, con $H$(m) y $Q$(m$^3$/s)

Desprecie las pérdidas singulares

### Solución

Balance de energía entre $A$ y $B$:

$$\begin{align}
B_{A} &= B_{B} + h_{T1} + h_{T2} + h_{\text{turb}_1} \\
500 &= 160 + R_{1}\cdot Q_{1}^{2} + R_{2}\cdot Q_{2}^{2} + h_{\text{turb}_1}
\end{align}$$

Así:

$$ h_{\text{turb}_1} = 250 - 50\cdot Q_{2}^{2} $$

Balance de energía entre $A$ y $C$:

$$\begin{align}
B_{A} &= B_{C} + h_{T1} + h_{T3} + h_{\text{turb}_2} \\
500 &= 110 + R_{1}\cdot Q_{1}^{2} + R_{3}\cdot Q_{3}^{2} + h_{\text{turb}_2}
\end{align}$$

Así:

$$ h_{\text{turb}_2} = 300 - 25\cdot Q_{3}^{2}  $$

Si ambas turbinas funcionan a la misma altura neta:

$$\begin{align}
h_{\text{turb}_1} &= h_{\text{turb}_2} \\
250 - 50\cdot Q_{2}^{2} &= 300 - 25\cdot Q_{3}^{2}
\end{align}$$

Además, es necesario considerar la continuidad del flujo:

$$\begin{align}
Q_{1} &= Q_{2} + Q_{3} \\
3 &= Q_{2} + Q_{3}
\end{align}$$

Así, tenemos 2 ecuaciones y dos incognitas, resolviendo:

$$ Q_{2} = 1 \text{ m}^{3}\text{/s} \qquad\qquad Q_{3} = 2 \text{ m}^{3}\text{/s} $$
