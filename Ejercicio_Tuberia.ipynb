Desde un depósito elevado se abastece de agua a un canal de regadío como se muestra en la figura. Calcule el flujo volumétrico del agua en el canal, si esta tiene 80ºF.

Viscosidad cinemática = 9,15*10-6

Rugosidad absoluta = 1,5*10-4


Solución

Comenzamos escribiendo la ecuación de energía (bernoulli). Se utilizan los puntos $A$ y $B$ como los puntos de referencia y simplificamos la ecuación tanto como sea posible. 

\begin{equation}
  \frac{p_{A}}{\gamma} + z_{A} + \frac{v_{A}^{2}}{2g} = \frac{p_{B}}{\gamma} + z_{B} + \frac{v_{B}^{2}}{2g} + h_{L}
\end{equation}

Como $p_{A} = p_{B} = v_{A} = 0$, entonces:

$$ z_{A} - z_{B} = \frac{v_{B}^{2}}{2g} + h_{L} \tag{1} $$

Si se analiza el problema, se puede observar que la corriente de agua en el punto $B$ tiene la misma velocidad que la del interior de la tubería. 

Se sabe que la diferencia de elevación $z_{A}-z_{B}$ es de 40 pies. Sin embargo, las pérdidas de energía dadas por $h_{L}$ dependen todas de la velocidad desconocida $v_{B}$. Así, se requiere iterar.

Exsiten cuatro componentes de la pérdida de energía total $h_{L}$:

$$ h_{L} = h_{1} + h_{2} + h_{3} + h_{4} $$

Donde:

$h_{1} = 1.0\frac{v_{B}^{2}}{2g} $ (Pérdida en la entrada)

$h_{2} = f\frac{L}{D} \frac{v_{B}^{2}}{2g} = f\frac{330}{0.3355} \frac{v_{B}^{2}}{2g} $ (Pérdida por frición en la tubería)

$ h_{3} =  f_{T}\frac{L_{e}}{D} \frac{v_{B}^{2}}{2g} = 20f_{T}\frac{v_{B}^{2}}{2g} $ (codo de radio largo)

$h_{4} = f_{T}\frac{L_{e}}{D} \frac{v_{B}^{2}}{2g} = 160f_{T}\frac{v_{B}^{2}}{2g} $ (válvula de compuerta abierta a la mitad)

De la tabla anterior, encontramos que para una tubería de acero de 4 pulgadas, $f_{T}= 0.017$. Entonces, tenemos:

$$ h_{L} = (1.0 + 985f + 20f_{T} + 160f_{T})\frac{v_{B}^{2}}{2g} = (4.06 + 985f)\frac{v_{B}^{2}}{2g} $$

Ahora sustituimos esta expresión en la ecuación (1) y despejamos $v_{B}$ en términos de $f$:

\begin{align}
z_{A} - z_{B} &= \frac{v_{B}^{2}}{2g} + h_{L}  \\
40\text{ pies} &= \frac{v_{B}^{2}}{2g} + (4.06 + 985f)\frac{v_{B}^{2}}{2g} \\
               &= (5.06 + 985f)\frac{v_{B}^{2}}{2g}
\end{align}

Al despejar $v_{B}$, queda:

$$ v_{B} = \sqrt{\frac{2g(40)}{5.06 + 985f}} = \sqrt{\frac{2580}{5.06 + 985f}} \tag{2}$$

El número de Reynolds y la rugosidad relativa vienen dados por:

$$Re = \frac{v_{B}D}{\nu} = \frac{v_{B}(0.3355)}{9.15\times 10^{-6}} = (0.366\times 10^{5})v_{B} \tag{3}$$

$$\frac{\varepsilon}{D} = 4.47\times 10^{-4}  $$

Esto represneta el comienzo del proceso de iteración, ¿Cuál es el rango posible de valores del factor de fricción para este sistema?

Debido a que $\varepsilon/D = 4.47\times 10^{-4}$, el valor más bajo posible de $f$ es 0.0155 para números de Reynolds muy altos, y el más alto posible es de 0.039 para un número de Reynolds igual a 4000. El valor inicial para el intento debe estar en este rango. Empleamos $f = 0.020$

Con las ecuaciones (2) y (3) encontramos los valores de la velocidad y el número de Reynolds:

$$ v_{B} =  \sqrt{\frac{2580}{5.06 + 985(0.02)}} = 10.2\text{ pies/s}$$

$$Re = (0.366\times 10^{5})(10.2) = 3.73\times 10^{5} $$

Para este número de Reynolds obtenido, en conjunto con la rugosidad relativa, volvemos a evaluar el factor de fricción en el diagrama de Moody y obtenemos un $f = 0.0175$. 

Con este nuevo valor de $f$, volvemos a calcular la velocidad y el número de Reynolds:

$$ v_{B} =  \sqrt{\frac{2580}{5.06 + 985(0.0175)}} = 10.8\text{ pies/s}$$

$$Re = (0.366\times 10^{5})(10.8) = 3.94\times 10^{5} $$

Con este nuevo Reynolds más la rugosidad relativa, evaluamos el factor de fricción en el diagrama de Moody. El valor encontrado es $f = 0.0175$, es decir, este valor no sufrió cambios, y el valor calculado para $v_{B}$ es correcto. Por lo tanto, tenemos:

$$ v_{B} = 10.8 \text{ pies/s} $$

$$ Q = A_{B}v_{B} = (0.0884\text{ pies}^{2})(10.8 \text{ pies/s}) = 0.955 \text{ pies}^{3}\text{/s} $$
